```python
import nmap
```

```python
nm = nmap.PortScanner()
ip = '82.137.166.100'
```

```python
nm.scan(ip)
```

```python
nm.scaninfo()
```

```python
cpes = []
for key in nm[ip]['tcp'].keys():
    if key == 80 or key == 8080:
        cpes.append([rec for rec in nm[ip]['tcp'][key]['cpe'].split(':')])
```

```python
i = 1
for cpe in cpes:
    print("{3}. Web server: {0} {1} {2}".format( cpe[2], cpe[3], cpe[4], i))
    i += 1
```
